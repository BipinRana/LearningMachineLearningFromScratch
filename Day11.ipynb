```python
from matplotlib import pyplot as plt
```

```python
company = ['A','B','C']
```

```python
x= [1,2,3]
```

```python
revenue = [10000,5000,7000]
```

```python
plt.bar(company,revenue,color=['r','b','g'],width=0.6,align='edge')
plt.title("Revenue by Companies")
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.grid()
plt.show()
```

```python
plt.barh(company,revenue,color=['r','b','g'],height=.5)
plt.title("Revenue by Companies")
plt.xlabel('Revenue')
plt.ylabel('Company')
plt.grid()
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
dead =  (df.Survived ==0).sum()
```

```python
survived = (df.Survived==1).sum()
```

```python
plt.bar(['Dead','Survived'],[dead,survived],color=['r','g'],width=0.6)
plt.title("Survival Graph")
plt.grid()
plt.show()
```
